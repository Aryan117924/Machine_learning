```python
import pandas as pd
df = pd.read_csv('/content/BMW_Car_Sales_Classification.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
categorical_cols = ['Model', 'Region', 'Color', 'Fuel_Type', 'Transmission', 'Sales_Classification']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
df
```

```python
x = df.drop(['Sales_Classification'], axis=1)
y = df['Sales_Classification']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#complement naive bais
from sklearn.naive_bayes import ComplementNB
clf = ComplementNB()
clf.fit(x, y)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy = clf.score(x_test, y_test)
print("Accuracy:", accuracy)
```

```python
#categorial naive bais
from sklearn.naive_bayes import CategoricalNB
clf = CategoricalNB()
clf.fit(x, y)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy = clf.score(x_test, y_test)
print("Accuracy:", accuracy)
```
